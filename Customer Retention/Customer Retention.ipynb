# Bivariate Analysis

```python
plt.figure(figsize=[18,15])

plt.subplot(2,2,1)
plt.title('Checking how long the customers shopping online on the basis of Gender')
sns.countplot(df['Shopping_Since'], hue=df['Gender'])

plt.subplot(2,2,2)
plt.title('Checking which device is used to access shopping online on the basis of Age')
sns.countplot(df['Device_Used'], hue=df['Age'])

plt.subplot(2,2,3)
plt.title('How the customers access online shopping & how many times they made purchase in 1 year')
sns.countplot(df['Internet_Accessibility'], hue=df['Shopping_Frequency'])

plt.subplot(2,2,4)
plt.title('In 1 year how many times customers made shopping & which city they shopped more')
sns.countplot(df['Shopping_Frequency'], hue=df['Shopping_City'])

plt.xticks(rotation=90)
plt.subplot(2,2,3).legend(loc='upper left',title='Shopping_Frequency')
plt.subplot(2,2,4).legend(loc='upper right',title='Shopping_City')
plt.show()
```

Observations:
1. Most of the female customers shoped online from more than 4 year and the count is also high for the females who shopped from 2-3 years.
2. Only few male customers shop online more than 4 years.
3. Many of the customers access the shopping websites more than 31-40 times in 1 year through Mobile Internet to shop the product
4. Most customers who used mobile internet to access the online shopping website made online purchase less than 10 times in a year, only few of the customers used wifi network to access the shopping store.
4. Most of the customers use ecommerce websites less than 10 times in a year from the city Delhi to shop the products.

```python
plt.figure(figsize=[18,13])

plt.subplot(2,2,1)
plt.title('Comparing screen size and the channel used to arrive at the online store',fontsize=12)
sns.countplot(df['Screen_Size'],hue=df['Channel_First_Used']);

plt.subplot(2,2,2)
plt.title('Comparing device used and how many times customers shopped in 1 year',fontsize=12)
sns.countplot(df['Device_Used'],hue=df['Shopping_Frequency']);

plt.subplot(2,2,3)
plt.title('Which OS and browser used to access the ecommerce website',fontsize=12)
sns.countplot(df['OS'],hue=df['Browser_Used']);

plt.subplot(2,2,4)
plt.title('How frequently the customers abandoned the shopping bags and why?',fontsize=12)
sns.countplot(df['Abandon_Frequency'],hue=df['Abandon_Reason']);

plt.subplot(2,2,2).legend(loc ="upper left",title="Shopping_Frequency");
plt.subplot(2,2,4).legend(loc ="upper right",title="Abandon_Reason");
plt.show()
```

Observations:
1. The customers having their mobile screen size say 5.5/ 6 inches(others) have followed search engine channel to arrive at their favorite online store for the first time.
2. Most of the customers use Smartphones 31-40 times in an year to access the ecommerce websites to shop the products.
3. Many customers having windows operating system in their device used Google chrome to access the ecommerce shopping websites and some of the customers having IOS/Mac operating system used Google chrome as well as Safari to reach the online shopping store.
4. Due to Lack of trust on the ecommerce websites, sometimes most of the customers abandoned the websites and some of the customers abandoned the shopping website because the promo code is not applicable, which means that most customers prefer discounts.
