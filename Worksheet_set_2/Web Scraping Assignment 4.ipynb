# Question 2

```python
driver = webdriver.Chrome(r"C:/Users/RAJIT/Downloads/chromedriver_win32/chromedriver.exe")
url = "https://www.bcci.tv/"
driver.get(url)

driver.maximize_window()

## Getting the url of International page:-
url_international = driver.find_element_by_xpath("//ul[@class='nav navbar-nav  align-items-center']/li[2]/a")
driver.get(url_international.get_attribute('href'))
```

```python
titles=[]
series=[]
place=[]
dates=[]
times=[]
while(True):
    
    ## Scrapping Match Title :
    title=driver.find_elements_by_xpath("//div[@class='tab-inner-content']/div/div/div/div/h5[2]/span")
    for i in title:
        titles.append(i.text)
  
   ## Scrapping Series:
    ser = driver.find_elements_by_xpath("//div[@class='fixture-card-bottom']/div")
    for i in ser:
        series.append(i.text)
    for i in series:
        place =i.split("-")[1]
        series_no =i.split("-")[0]  
    
    ## Scrapping Date
    date = driver.find_elements_by_xpath("//div[@class='match-card-left match-schedule']/h5")
    for i in date:
        dates.append(i.text)
    
     ## Scrapping Time:
    time = driver.find_elements_by_xpath("//div[@class='match-card-right match-schedule ']/h5")
    for i in time:
        times.append(i.text)
       
    try:
        more_fixture = driver.find_element_by_xpath("//div[@class='col-lg-12 col-md-12 col-sm-12']/button")
        
        more_fixture.click()
        time.sleep(5)
    except:
        break

df = pd.DataFrame({'Match Title':titles,
                         'Series':series_no ,
                         'Place' :place,
                         'Date':dates,
                         'Time':times
    
                        })
df
```
